# Data Collection, Preprocessing, and Exploration

### Objective: Collect, preprocess, and explore datasets suitable for classification and detection.

## Pipeline
| Member | Assigned Tasks | Description |
| --- | --- | --- |
| 1️⃣ | **Data sourcing & download** | Gather datasets (RSNA Breast Cancer + auxiliary datasets like COCO/VOC for detection). Ensure correct labeling and file structure. |
| 2️⃣ | **Image preprocessing pipeline** | Resize, normalize, and standardize images; handle corrupt files; create preprocessing functions. |
| 3️⃣ | **Data augmentation** | Apply transformations: rotation, flipping, brightness jitter, etc. Document each method. |
| 4️⃣ | **Train/Val/Test splitting** | Implement stratified splitting to handle class imbalance (especially in cancer data). |
| 5️⃣ | **EDA visualization & documentation** | Visualize class distribution, example images, and bounding boxes. Identify dataset biases. Compile preprocessing + EDA into one notebook and produce a short report. |
